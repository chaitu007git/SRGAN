## Deleted Image Deduplication Script Using Hashing (MD5/SHA256)








```python
import os
import hashlib
from PIL import Image

def get_image_hash(image_path, hash_algorithm="md5"):
    """
    Calculate the hash of an image using MD5 or SHA256 algorithms.
    """
    try:
        with Image.open(image_path) as img:
            img = img.convert("RGB")  # Convert the image to RGB format
            img = img.resize((8, 8))  # Resize the image to speed up hashing
            pixels = list(img.getdata())
            avg_pixel = sum(sum(pixel) for pixel in pixels) // len(pixels)  # Correctly calculate the average pixel color

            hash_value = 0
            for pixel in pixels:
                # Calculate the hash value based on the average
                hash_value = (hash_value << 1) | (1 if sum(pixel) > avg_pixel else 0)
            
            if hash_algorithm == "md5":
                return hashlib.md5(str(hash_value).encode('utf-8')).hexdigest()
            elif hash_algorithm == "sha256":
                return hashlib.sha256(str(hash_value).encode('utf-8')).hexdigest()
            else:
                raise ValueError("Invalid hash algorithm provided.")
    except Exception as e:
        print(f"Error processing the image {image_path}: {e}")
        return None

def remove_duplicates(directory, hash_algorithm="md5"):
    """
    Remove duplicate images from the specified directory.
    """
    hashes = {}
    duplicates = []
    
    # Iterate through all files in the directory
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            if file_path.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                image_hash = get_image_hash(file_path, hash_algorithm)
                if image_hash:
                    if image_hash in hashes:
                        duplicates.append(file_path)
                        os.remove(file_path)  # Delete the duplicate image
                        print(f"Deleted duplicate image: {file_path}")
                    else:
                        hashes[image_hash] = file_path

    print(f"\nFound and deleted {len(duplicates)} duplicate images.")

# Test the code
directory = r"D:\2-SRGAN for Image\Second Train\Data\New Data don't use"  # Replace with the actual folder path
remove_duplicates(directory, hash_algorithm="md5")
```

## Image Resizing Script for 1024x1024 Resolution with Small Image Filtering

```python
import os
from PIL import Image

# Define input and output folders
input_folder = r"D:\2-SRGAN for Image\Second Train\Streamlit\SRGAN_Streamlit\data\Test_Image"  # Specify the input folder path
output_folder = r"D:\2-SRGAN for Image\Second Train\Streamlit\SRGAN_Streamlit\data\Test_Image"  # Specify the output folder path

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get a list of all image files in the input folder
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Process each image in the input folder
for file_name in image_files:
    file_path = os.path.join(input_folder, file_name)

    try:
        # Open the image
        with Image.open(file_path) as img:
            # Get image dimensions
            width, height = img.size

            # Check if the image resolution is 1024x1024 or larger
            if width >= 1024 and height >= 1024:
                # Resize the image to 1024x1024
                img_resized = img.resize((256, 256))
                
                # Define the new path to save the resized image
                output_path = os.path.join(output_folder, file_name)
                
                # Save the resized image
                img_resized.save(output_path)
                print(f"Resized image {file_name} saved to {output_path}")
            else:
                # Skip or delete images smaller than 1024x1024
                print(f"Image {file_name} ignored because it is smaller than 1024x1024.")
    
    except Exception as e:
        print(f"Error processing image {file_name}: {e}")

print("Processing completed successfully.")
```

## Shuffle and Rename Images with Sequential Numbering

```python
import os
import random

# Path to the folder containing images
input_folder = r"D:\2-SRGAN for Image\Second Train\Data\New Data don't use"  # Replace with your folder path

# Collect all files with supported image extensions
image_files = [file for file in os.listdir(input_folder) if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Shuffle the list of images randomly
random.shuffle(image_files)

# Rename images with sequential numbering starting from 2050
for new_index, file_name in enumerate(image_files, start=0):
    # Original image path
    source_path = os.path.join(input_folder, file_name)
    
    # New image name and path
    new_file_name = f"{new_index}.jpg"  # Save all images with a unified extension (.jpg)
    destination_path = os.path.join(input_folder, new_file_name)
    
    # Rename the image
    os.rename(source_path, destination_path)

print(f"Images have been renamed with sequential numbering starting from 2050 in {input_folder}.")
```

## check if image resolution in Folder is 1024*1024

```python
import os
from PIL import Image

def check_image_resolution(directory, required_width=1024, required_height=1024):
    """
    Check if all images in the specified directory have the required resolution.

    Parameters:
        directory (str): Path to the folder containing images.
        required_width (int): Required image width. Default is 1024.
        required_height (int): Required image height. Default is 1024.

    Returns:
        None
    """
    # Collect all files with supported image extensions
    image_files = [file for file in os.listdir(directory) if file.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    # Flag to track if all images meet the resolution
    all_correct = True

    for file_name in image_files:
        file_path = os.path.join(directory, file_name)
        try:
            with Image.open(file_path) as img:
                width, height = img.size
                if width != required_width or height != required_height:
                    print(f"Image '{file_name}' has resolution {width}x{height}, which does not match {required_width}x{required_height}.")
                    all_correct = False
        except Exception as e:
            print(f"Error processing image '{file_name}': {e}")

    if all_correct:
        print(f"All images in the folder '{directory}' have the resolution {required_width}x{required_height}.")
    else:
        print(f"Some images in the folder '{directory}' do not meet the required resolution.")

# Specify the folder path
folder_path = r"D:\2-SRGAN for Image\Second Train\Data\Data-1024"  # Replace with the path to your folder
check_image_resolution(folder_path)
```
